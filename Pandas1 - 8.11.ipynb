```python
train["Nationality"] = "Unknown"
train.loc[train["Embarked"].isnull(), "Embarked"] = "Unknown"


train["Embarked"] == "C"

# train.loc[train["Embarked"]]

# if (train.loc[train["Embarked"]]) == ("C"):
#     train["Nationality"] = "France"
# elif (train.loc[train["Embarked"]]) == ("S" | "Q"):
#     train["Nationality"] = "UK"
# else:
#     train["Nationality"] = "Unknown"

# train[["Embarked", "Nationality"]].head()
```

#### 함수 사용 (python 문법 사용가능)

```python
def find_nationality(embarked):
    if embarked == "C":
        return "France"
    elif embarked == "S" or embarked == "Q":
        return "UK"
    else:
        return "Unknown"
    
# train[["Embarked","Nationality"]].head()

train["Embarked"].apply(find_nationality)
```
