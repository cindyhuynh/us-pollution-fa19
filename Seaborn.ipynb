```python
import pandas as pd
```

```python
train = pd.read_csv("~/Downloads/all/train.csv", index_col="PassengerId")

print(train.shape)
train.head()
```

# Explore

```python
%matplotlib inline

import seaborn as sns
```

## 1. Countplot
#### x축만

```python
sns.countplot(data=train, x="Embarked", hue="Survived")
```

```python
sns.countplot(data=train, x="Pclass", hue="Survived")
```

## 2. Barplot - 숫자데이터에 적합
#### x축 y축

```python
sns.barplot(data=train, x="Pclass", y="Fare", hue="Survived")
```

## 3. Pointplot
barplot과 비슷, x축 순서상 연관성이 있을 때

```python
sns.pointplot(data=train, x="Pclass", y="Fare", hue="Survived")
```

## 4. Lmplot

fit_reg = true : 선이 하나 나오는데 별 필요없으면 false로 지우고 사용


```python
sns.lmplot(data=train, x="Age", y="Fare", hue="Survived", fit_reg=False)
```

outlier들 제거하여 좀더 구체적으로 시각화

```python
low_fare = train[train["Fare"] < 100 ]

print(train.shape, low_fare.shape)

sns.lmplot(data=low_fare, x="Age", y="Fare", hue="Survived", fit_reg=False)
```

## 5. Distplot
distribution plot

가우시안 그래프 expect -> 아닐 시 그 부분을 파고들기

```python
# sns.distlpt(data=train, x="Fare")  --> XXX

sns.distplot(train["Fare"])
```

```python
sns.distplot(low_fare["Fare"], hist = False)
```
